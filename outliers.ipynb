```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Sample data with an outlier
data = np.array([250, 275, 280, 300, 320, 325, 330, 350, 360, 10000])

# Calculate the Z-scores for each data point
z_scores = np.abs(stats.zscore(data))

# Define a threshold for identifying outliers (commonly, a threshold of 3 is used)
threshold = 3

# Find and print the outliers based on the Z-score
outliers = np.where(z_scores > threshold)[0]
print("Outliers (Indices):", outliers)

# Create an array to specify the color of each point
colors = "red"

# Visualize the data with outliers highlighted
plt.figure(figsize=(8, 6))
plt.scatter(np.arange(len(data)), data, c=colors)
plt.xlabel("Data Point Index")
plt.ylabel("Sale Price (in $000s)")
plt.title("Outliers Detection using Z-score")
plt.show()
```
